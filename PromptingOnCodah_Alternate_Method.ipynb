```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def encode_sentence(sentence):
    inputs = tokenizer_bert(sentence, return_tensors='pt', truncation=True, padding=True)
    inputs = {key: val.to(device) for key, val in inputs.items()}  # Move inputs to GPU if available
    with torch.no_grad():
        outputs = model_bert(**inputs)
    # Use the [CLS] token's output as the sentence embedding
    return outputs.last_hidden_state[:, 0, :].detach().cpu().numpy()  # Move tensor to CPU before converting to NumPy

def pick_best_sentence_label(sentences):
    # Encode each sentence
    embeddings = np.vstack([encode_sentence(sentence) for sentence in sentences])

    # Compute cosine similarities between all sentence embeddings
    similarities = cosine_similarity(embeddings)

    # Sum similarities for each sentence to use as a heuristic for "best" sentence
    scores = similarities.sum(axis=1)

    # Select the index of the highest scoring sentence
    best_index = np.argmax(scores)

    # Return 1-based index
    return best_index + 1

def pick_best_sentence_labels(lists_of_sentences):
    best_labels = []
    for sentences in lists_of_sentences:
        best_label = pick_best_sentence_label(sentences)
        best_labels.append(best_label)
    return best_labels


# Get the best sentence labels for each list
for i, phrase_list in enumerate(start_test):
    phrase = phrase_list[0]
    #print(f"Given the start phrase:\n{phrase}")
    #print("Choose the best option below:")

    for choice in choices_codah_test[i]:
      indexed_choices_text = "\n".join(choices_codah[i])
      #print(choice)

    #print("\nThe best option is:\n")
    prompt = f"Given the start phrase:\n{phrase}\n\nChoose the best option below:\n{indexed_choices_text}\n\nThe best option is:"
    print(prompt)

best_labels = pick_best_sentence_labels(prompt)
predicted_answer_BERT_codah  = []
print("Best Sentence Labels:")
for idx, best_label in enumerate(best_labels):
    predicted_answer_BERT_codah.append(best_label)
    print(f"Best sentence label from list {idx + 1}: {best_label}")
```

```python
def get_best_choice_index_BERT(prompt, choices):
    scores = []
    for idx, choice in enumerate(choices):
        # Combine prompt and choice
        input_text = f"{prompt} {choice}"

        # Encode the combined text
        inputs = tokenizer_BERT(input_text, return_tensors='pt').to(device)

        # Calculate token probabilities
        with torch.no_grad():
            outputs = model_BERT(**inputs)
            logits = outputs.logits

        # Get the input_ids of the choice part
        choice_ids = tokenizer_BERT.encode(choice, add_special_tokens=False, return_tensors='pt').to(device)
        choice_len = choice_ids.size(1)
        # Extract the logits corresponding to the choice part
        choice_logits = logits[0, -choice_len:, :]
        choice_probs = torch.softmax(choice_logits, dim=-1)
        choice_log_probs = torch.log(choice_probs)

        # Compute the total log probability for the choice
        total_log_prob = 0
        for i in range(choice_len):
            total_log_prob += choice_log_probs[i, choice_ids[0, i]].item()

        scores.append((idx, total_log_prob))

    # Find the index of the choice with the highest log probability
    best_choice_idx = max(scores, key=lambda x: x[1])[0]
    return best_choice_idx
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_bart = model_bart.to(device)
```

```python
def calculate_score(prompt, choice):
    # Formulate the complete prompt
    full_prompt = f"{prompt} {choice}"

    # Tokenize the input
    input_ids = tokenizer_bart.encode(full_prompt, return_tensors='pt').to(device)

    # Generate output with BART
    outputs = model_bart(input_ids, labels=input_ids)

    # Get the loss (negative log likelihood)
    loss = outputs.loss.item()

    return -loss  # Higher score is better

def get_best_choice_index_for_prompt(prompt, choices):
    best_choice_index = -1
    highest_score = float('-inf')

    for i, choice in enumerate(choices):
        # Calculate the score for each choice
        score = calculate_score(prompt, choice)

        # Select the choice with the highest score
        if score > highest_score:
            highest_score = score
            best_choice_index = i

    return best_choice_index

def get_best_choice_indices(prompts, all_choices):
    best_indices = []
    for prompt, choices in zip(prompts, all_choices):
        best_index = get_best_choice_index_for_prompt(prompt, choices)
        best_indices.append(best_index)
    return best_indices

best_indices = get_best_choice_indices(startPhrase_swag, choices_swag)

print("Best Choice Indices:")
predicted_answer_bart_swag = []
for prompt, best_index in zip(startPhrase_swag, best_indices):
    print(f"Best choice index for '{prompt}': {best_index}")
    predicted_answer_bart_swag.append(best_index)
    #print(f"Best choice: {choices[startPhrase_swag.index(prompt)][best_index]}")
```

```python
def format_mcq_prompt(question, choices):
    prompt = f"Question: {question}\n"
    for i, choice in enumerate(choices):
        # prompt += f"{chr(65 + i)}. {choice}\n"
        prompt += f"{i}: {choice}\n"
        # prompt += f"{choice}\n"
    prompt += "Answer:"
    return prompt

def generate_answer(prompt, model, tokenizer, max_length=256):
    inputs = tokenizer(prompt, return_tensors='pt').to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=max_length,eos_token_id=50256, pad_token_id=50256)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer.strip()
```

```python
# Define a function to calculate log-probabilities of a given completion
def calculate_log_prob(prompt, completion, model, tokenizer):
    # Combine prompt and completion
    text = prompt + completion
    # Tokenize the combined text
    input_ids = tokenizer.encode(text, return_tensors='pt').to(device)
    # Calculate the model's output logits
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
    # Get the loss (negative log-likelihood)
    log_prob = -outputs.loss.item()
    return log_prob

# Define a function to find the best completion
def find_best_completion(prompt, choices, model, tokenizer):
    best_choice = None
    best_log_prob = float('-inf')

    for choice in choices:
        log_prob = calculate_log_prob(prompt, choice, model, tokenizer)
        if log_prob > best_log_prob:
            best_log_prob = log_prob
            best_choice = choice

    return best_choice
```
