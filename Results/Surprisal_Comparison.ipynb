```python
def accuracy(predicted_labels, true_labels):
  return accuracy_score(predicted_labels, true_labels)
```

```python
def agreement(predicted_labels_1, predicted_labels_2):
  if len(predicted_labels_1) != len(predicted_labels_2):
    print("Lists are different length")
    return
  agreement = 0
  for i in len(predicted_labels_1):
    if predicted_labels_1[i] == predicted_labels_2[i]:
      agreement+=1
  return agreement/len(predicted_labels_1)
```
